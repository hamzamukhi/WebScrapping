```python
import requests
from bs4 import BeautifulSoup
url = 'https://bagstore.pk'
```

```python
# Step 1 get the html 
r = requests.get(url)
htmlContent = r.content
print(htmlContent)
```

```python
#step 2 Parse the Html
soup = BeautifulSoup(htmlContent, 'html.parser')
print (soup.prettify)
```

```python
#step 3 Html Tree Traversal

#commonly used types of objects:
#1. Tag
#2. NavigableString
#4. BeautifulSoup
#5. Comment

#Get the title of the html page
title = soup.title
print(title)

print(type(soup))
print(type(title))
print(type(title.string))
```

```python
#Get all the paragraphs from the page
paras = soup.find_all('p')
print(paras)
```

```python
#Get all the anchors from the page
anchors = soup.find_all('a')
print(anchors)
```

```python
#get first element in the html page
print(soup.find('p'))

#print(soup.find('p')["class"])  #there is no class

#find all the elements with class lead
#print(soup.find_all("p",class_="lead"))
```

```python
#get the text from tags / soup
print(soup.find('p').get_text())
print(soup.get_text())
```

```python

#get all the anchor tags from the page
anchors = soup.find_all('a')
all_links = set()
for link in anchors:
  if(link.get('href') != '#'):
    linkText = "https://bagstore.pk" +link.get('href')
    all_links.add(link)
    print(linkText)
```
